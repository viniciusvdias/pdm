```python
import pandas as pd
import matplotlib.pyplot as plt

print(plt.__name__)
print(pd.__name__)
```

```python
df = pd.read_csv("times_12.csv")
```

```python
print(df)
```

```python
fig, ax = plt.subplots(figsize=(14, 8))
code_sessions = df['code_session'].unique()

for session in code_sessions:
    session_df = df[df['code_session'] == session]
    ax.plot(session_df['amount of files'], session_df['time'], marker='o', linestyle='-', label=session)

ax.set_xlabel('Quantidade de Arquivos', fontsize=12)
ax.set_ylabel('Tempo de Execuo (s)', fontsize=12)
ax.set_title('Progresso do Tempo de Execuo por Etapa do Cdigo', fontsize=16)

ax.legend(title='Etapa do Cdigo')

ax.grid(True, which='both', linestyle='--', linewidth=0.5)
```
